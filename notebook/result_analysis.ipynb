```python
import json
import pandas as pd
import os
```

```python
DATA_ROOT = "/data/joe/misinformation/FRESH/output/{}/bert_classification/fresh/wrapper_saliency/{}_thresholder/{}/human_supervision=0/model_b/"
dataset = ['pheme_all', 'pheme_drop_unverified']
thresholder = ['top_k', 'contiguous']
ratio = ['0.3', '0.5']
```

# Model with all input text

```python
all_input = "/data/joe/misinformation/FRESH/output/{}/bert_classification/fresh/metrics.json"
results = []
for data in dataset:
    file = all_input.format(data)
    result = json.load(open(all_input.format(data)))
    results.append(
        {
            "name":data,
            "f1":result["test_macro_fscore"],
            "precision":result["test__macro_precision"],
            "recall":result["test__macro_recall"],
        }
    )
pd.DataFrame(results)
```

# Model with selected rationales

```python
predictor_ROOT = "/data/joe/misinformation/FRESH/output/{}/bert_classification/fresh/wrapper_saliency/{}_thresholder/{}/human_supervision=0/model_b/metrics.json"
results = []
for data in dataset:
    for th in thresholder:
        for ra in ratio:
            file = predictor_ROOT.format(data, th, ra)
            result = json.load(open(file))
            results.append(
                {
                    "name":"-".join([data, th, ra]),
            "f1":result["test_macro_fscore"],
            "precision":result["test__macro_precision"],
            "recall":result["test__macro_recall"],
                }
            )
pd.DataFrame(results)
```

# Rationale extraction performance

```python
predictor_ROOT = "/data/joe/misinformation/FRESH/output/{}/bert_classification/fresh/wrapper_saliency/{}_thresholder/{}/human_supervision=0/metrics.json"
results = []
for data in dataset:
    for th in thresholder:
        for ra in ratio:
            file = predictor_ROOT.format(data, th, ra)
            result = json.load(open(file))
            results.append(
                {
                    "name":"-".join([data, th, ra]),
                    "f1":result["test_f1"],
                    "precision":result["test_p"],
                    "recall":result["test_r"],
                }
            )
pd.DataFrame(results)
```
